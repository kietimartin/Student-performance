```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# handling errors 
try:
    data = pd.read_csv("student.csv")
    df = pd.DataFrame(data)

except(FileNotFoundError):
    print('Check your file and try again') 
```

```python
# printing the first few columns
df.head()
```

```python
# getting the summary statistics of the columns with digit values
df.describe()
```

```python
# Getting the shape of the dataframe
df.shape
```

```python
# getting information about the columns and their values to check for missing values
df.info()
```

```python
# checking the number of missing values
df.isna().sum()
```

```python
g = sns.catplot(x="final_grade", data=df, kind="bar", col="school_support")

g.fig.suptitle("Final Grade Distribution by School Support", y=1.02)

g.set_titles("School Support: {col_name}")

plt.show()
# Data shows that despite the differences in those receiving support and those that do not the performance is almost similar.
```

```python
g = sns.relplot(x="study_hours", y="final_grade", data=df, kind="scatter")
# adding the plots title
g.fig.suptitle("Relationship Between Study Hours and Final Grade", y=1.02)
plt.show()
# the scatter plot shows there is no relationship between the number of study hours and final grade
```

```python
plt.hist(x="final_grade", bins=20, data=df)
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.title("Distribution of Final Grades")
plt.show()
# The distribution doesn't seem strongly skewed to the left (negatively skewed) or the right (positively skewed)
```

```python
g = sns.catplot(x='school_support', data=df, kind='count', order=['yes', 'no'])

g.fig.suptitle("Number of Students by School Support", y=1.05)
g.set_ylabels("Number of Students")
plt.show()
# the number of students who received school support are slightly more than those who did not receive support.
```
